```python
a = float(input("a: "))
b = float(input("b: "))
c = float(input("c: "))
x1 = (-b + (b ** 2 - 4 * a * c) ** 0.5) / (2 * a)
x2 = (-b - (b ** 2 - 4 * a * c) ** 0.5) / (2 * a)
print("x1 = ", x1)
print("x2 = ", x2)
```

```python
product = 1
first = 1
while first <= 5:
    num = float(input("Enter a number: "))
    product = product * num
    first = first + 1
print("The product is: ", product)
```

```python
num = int(input("Bitte geben Sie eine Zahl ein: "))
num_r = str(num) [::-1]
print(str(num), num_r)
if str(num) == num_r:
    print("Ja")
if str(num) != num_r:
    print("Nein")
```
